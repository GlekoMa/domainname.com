## TOPSIS法求解

TOPSIS法也称为理想解法，是一种有效的多指标评价方法。

理想解法相关公式有以下：
$$
c_j^*=\begin{cases}
\max c_{ij},\ \ j\text{为效益型属性}\\
\min c_{ij},\ \ j\text{为成本型属性}.
\end{cases}
$$ {#eq-正理想解}

· $c_j^*$为第$j$列的正理想解（向量）。

$$
c_j^0=\begin{cases}
\min c_{ij},\ \ j\text{为效益型属性}\\
\max c_{ij},\ \ j\text{为成本型属性}.
\end{cases}
$$ {#eq-负理想解}

· $c_j^0$为第$j$列的负理想解。
